```python
import pandas as pd
import detect
import detect_distance_calc
import os
```

```python
from PIL import Image, ImageEnhance
import os
import numpy as np
# set input and output directories
work_dir = "/content/drive/MyDrive/BacteriaDetection/YoloV5L/images" #Make sure this location is correct
for dirname,sub_dir,filenames in os.walk(work_dir):
  for filename in filenames:
    # Load TIFF image
    img = Image.open(os.path.join(work_dir,filename))

    # # Convert to Numpy array
    arr = np.array(img)

    # # Normalize to 0-255 range
    arr = (arr - arr.min()) * (255 / (arr.max() - arr.min()))

    # # Convert to 8-bit grayscale
    arr = arr.astype(np.uint8)
    img = Image.fromarray(arr, mode="L")

    # # Save as PNG
    img.save('/content/drive/MyDrive/BacteriaDetection/YoloV5L/images/pngs/'+filename[:-4]+'.png') #Make sure this location is correct
```

```python
data = []
work_dir = '/content/drive/MyDrive/BacteriaDetection/YoloV5L/images/pngs/' #Make sure this location is correct
for dirname,_,filenames in os.walk(work_dir):
  for filename in filenames:
    opt = {'weights':'/content/drive/MyDrive/BacteriaDetection/YoloV5L/weights/best.pt', #Make sure this location is correct
           'imgsz':(512,512),
           'conf_thres':0.25,
           'source':work_dir+'/'+filename,
           'project':'/content/drive/MyDrive/BacteriaDetection/YoloV5L/images/detect', #Make sure this location is correct
           'exist_ok':True}
    preds,distances,exit_idx = detect_distance_calc.run(**opt)
    set_of_levels,dP_Level = filename[:-4].split('-')[:2]
    set_of_levels = int(set_of_levels)
    dP_Level = int(dP_Level)
    for distance,idx in list(zip(distances,exit_idx)):
      data.append([set_of_levels,dP_Level,distance,idx])

data.sort()
df = pd.DataFrame(columns=['Set','dP','Distance Traveled (um),ChannelID'],data=data)

groups = df.groupby(['Set','dP','ChannelID'])
dist = []
single_cell = []
for group in (groups.groups):
  g = groups.get_group(group)
  g = g.sort_values(by='Distance Traveled (um)', ascending=False)
  dist.append(g['Distance Traveled (um)'].values[0])
  if len(g) > 1:
      single_cell.append(0)
  else:
      single_cell.append(1)

df_reduced = groups.first().reset_index()
df_reduced['Distance Traveled (um)'] = dist
df_reduced['Single Cell'] = single_cell
df_reduced.to_csv('/content/drive/MyDrive/BacteriaDetection/01312023.csv',index=False) #Make sure this path is correct
```
